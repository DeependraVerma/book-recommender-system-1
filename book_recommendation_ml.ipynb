## Popularity Based Recommendation System

```python
rating.head()
```

```python
books.head()
```

```python
books_with_rating = rating.merge(books, on =  "ISBN")
```

```python
books_with_rating.head()
```

```python
num_rating = books_with_rating.groupby("Book-Title").count()["Book-Rating"].reset_index()
```

```python
num_rating.rename(columns = {"Book-Rating":"Num Rating"}, inplace=True)
```

```python
num_rating
```

```python
avg_rating = books_with_rating.groupby("Book-Title").mean("Book-Rating")["Book-Rating"].reset_index()
```

```python
avg_rating.rename(columns={"Book-Rating":"Avg_Rating"}, inplace=True)
```

```python
popular_df = num_rating.merge(avg_rating,on = "Book-Title")
```

```python
popular_df = popular_df[popular_df["Num Rating"]>=250].sort_values(by = "Avg_Rating",ascending = False)
```

```python
popular_df = popular_df.head(50)
```

```python
popular_df
```

```python
popular_df = popular_df.merge(books, on = "Book-Title")
```

```python
popular_df.drop_duplicates("Book-Title",inplace=True)
```

```python
popular_df = popular_df[["Book-Title","ISBN","Image-URL-M","Book-Author","Num Rating","Avg_Rating"]]
```

```python
popular_df
```

## Colaborative Recommendation System

```python
books_with_rating
```

```python
user_200 = books_with_rating.groupby("User-ID").size().reset_index()
```

```python
user_200
```

```python
user_200 = user_200[user_200[0]>200]
```

```python
user_200
```

```python
book_50 = books_with_rating.groupby("Book-Title").size().reset_index()
```

```python
book_50 = book_50[book_50[0]>=50]
```

```python
book_50
```

```python
user_ = books_with_rating.merge(user_200, on = "User-ID")
```

```python
user_
```

```python
user_book = user_.merge(book_50, on = "Book-Title")
```

```python
user_book.drop_duplicates()
```

```python
pt = user_book.pivot_table(index="Book-Title", columns= "User-ID",values = "Book-Rating")
```

```python
pt.fillna(0, inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score
```

```python
similarity_score[0]
```

```python
def recommend(book_name):
    index = np.where(pt.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])),key = lambda x:x[1],reverse=True)[1:6]
    
    for i in similar_items:
        print(pt.index[i[0]])
```
